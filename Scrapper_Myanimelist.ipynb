```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from tqdm import tqdm
from glob import glob
import requests
import time
import os
```

```python
options = webdriver.ChromeOptions()
options.add_argument("-headless")
options.add_argument("-no-sandbox")
options.add_argument("-disable-dev-shm-usage")
wd = webdriver.Chrome("chromedriver",options=options)
```

```python
wt = WebDriverWait(wd, 2)
```

```python
links = []
```

100   in (100, 200, 50)

##Scraping
⚛ Para 10000 datos: for pg in range(200)
⚛ Para otro 10000 datos: for pg in range(200,400)

```python
for pg in range(500,600):
  wd.get("https://myanimelist.net/topanime.php?limit=" + str(pg))
  link = wd.find_elements(By.CSS_SELECTOR, 'div[class="detail"] h3 a')
  for index in link:
    links.append(index.get_attribute('href'))
```

```python
len(links)
```

```python
links
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
d_titulo = []
d_synopsis = []
```

```python
import time
```

```python
inicio = time.time()
for index_link in links:
  try:
    wd.get(index_link)
    titulo = wt.until(ec.visibility_of_element_located((By.CSS_SELECTOR, 'div[class="h1-title"]')))
    synopsis = wt.until(ec.visibility_of_element_located((By.CSS_SELECTOR,'p[itemprop*="description"]')))

    d_titulo.append(titulo.text)
    d_synopsis.append(synopsis.text)
    print(index_link)
  except NoSuchElementException as e:
    print('exepctions...')

fin = time.time()
print(fin-inicio)
```
